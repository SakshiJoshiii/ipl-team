coverting the hs data type and cleaning the data

```python
def clean(year):

  year['HS'] = year['HS'].str.replace('*', '')
  year['HS'] = year['HS'].astype(int)

  year['Avg'] = year['Avg'].astype(str)
  year['Avg'] = year['Avg'].str.replace('-', '0')
  year['Avg'] = year['Avg'].astype(float)

  return year

bat = clean(bat)

bat16 = clean(bat16)
bat17 = clean(bat17)
bat18 = clean(bat18)
bat19 = clean(bat19)
bat20 = clean(bat20)
bat21 = clean(bat21)
bat22 = clean(bat22)
```

Murge the data on the basis of seasons


```python
batting_list = [bat16,bat17,bat18,bat19,bat20,bat21,bat22]
def combineddata(list_data):
  #first of all we check the no of season played by each player
  season_count = pd.concat(list_data)['Player'].value_counts().reset_index()
  #to rename the column name
  season_count = season_count.rename(columns={'Player':'Seasons','index':'Player'})
  #to group the data on the basis of players
  alldata = pd.concat(list_data).groupby(['Player']).sum().reset_index()
  # to merge the season_count with alldata
  all_data = alldata.merge(season_count, on='Player',how='left')
  return all_data

bat = combineddata(batting_list)
bat
```

Finding avg and sr on the basis of seasons

```python
def filter(all_data):
  all_data['Avg'] = round(all_data['Avg'] / all_data['Seasons'], 2)
  all_data['SR'] = round(all_data['SR'] / all_data['Seasons'], 2)
  return all_data

bat = filter(bat)
bat

```

```python
list_data = [ball16, ball17, ball18, ball19, ball20, ball21, ball22]
def combine(data):
  season = pd.concat(list_data)['Player'].value_counts().reset_index()
  season = season.rename(columns = {'Player':'Seasons','index':'Player'})
  alldata = pd.concat(data).groupby(['Player']).sum().reset_index()
  alldata = alldata.merge(season, on = 'Player', how = 'left')
  return alldata

ball = combine(list_data)
ball
```

```python
def converting(df_data):
  df_data['Avg'] = round(df_data['Avg'] / df_data['Seasons'],2)
  df_data['SR'] = round(df_data['SR'] / df_data['Seasons'],2)
  return df_data

ball = converting(ball)
```
